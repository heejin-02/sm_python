5. 전년대비 범죄 발생 횟수 증감률 계산
- 공식 : (금년 - 작년) / 작년 * 100

```python
# 15년 ~ 16년 사이의 범죄 증감률 15년 ~ 16년 사이의 범죄 증감률
s1516 = (s16 - s15) / s15 * 100

# 16~ 17년
s1617 =  (s17 - s16) / s16 * 100
```

```python
s1617
```

변수들 정리
- s15 : 15년 총계
- s16 : 16년 총계
- s17 : 17년도 총계
- s1516 : 15~16 증감률
- s1617 : 16~17 증감률
