```python
def s_replace(s):
    result = '' # result   
    for i in s:
         # s  ,   
            # result       
            #   ''  
        if i != '':#  '' 
            result += i # result    
    return result
```

```python
num1 = int(input("    >> "))
num2 = int(input("    >> "))
op = input(" (+,-) >> ")
result = cal(num1, num2, op)
print(f" : {result}")
```

```python
def cal(num1, num2, op):
    if op == '+':
        return num1+num2
    elif op == '-':
       return num1 - num2
    else:
       return " "
```

```python
pop = pd.Series([9600000,3340000, 1480000, 2410000],
               index = ['','','','']
               )
pop
```

```python
pop + 100
```

```python
#  100   
def plus_100(x):
    return x+100
```

```python
# apply  
# Series   
pop.apply(plus_100)
```

```python
#   pop 
plus_100(pop)

# pop  
# return : pop + 100
```

```python
# 
# 1. plus_100(pop) : pop   ()  ->  : pop + 100
# 2. pop.apply(plus_100) :   pop   
# -> 9600000 + 100, 3340000 + 100 ... 
```

```python
# 300    (  0 )
def con_return(x):
    if x >= 3000000:
        return x
    else:
        return 0
```

```python
# apply  
pop.apply(con_return)
```

```python
# apply 
#   pop  
con_return(pop)
```
