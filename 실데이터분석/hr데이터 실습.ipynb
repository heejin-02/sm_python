#### org_info 데이터(조직정보)에 각 조직별 조직 인원수, 평균나이, 평균근무년수, 평균연봉

```python
# 조직정보 데이터 확인
org_info = pd.read_csv('org-info.csv', encoding='euc-kr', index_col='ORG_CD')
org_info.head()
```

```python
# 인사정보 데이터 확인
hr_info = pd.read_csv('hr-info.csv', encoding='euc-kr')
hr_info
```

```python
# 조직별 인원수
hr_info.groupby('ORG_CD').count()

# 조직정보 데이터와 공통적으로 갖고 있는 컬럼 ORG_CD
# ORG_CD 컬럼을 기준으로 groupby
```

```python
# hr_info 에서 groupby를 이용해 조직별 조직 인원수, 평균나이, 평균근무년수, 평균연봉
org_agg = hr_info.groupby('ORG_CD').agg({'ENAME' : 'count',
                                         'AGE' : 'mean',
                                         'SERVICE_YEAR' : 'mean',
                                         'SALARY' : 'mean' })
org_agg
```

```python
org_info.head(1)
```

```python
# org_info와 조직구성원 데이터에서 추출한 df을 열방향으로 병합
# 주의 : concat을 이용한 열방향 병합시, 각 DataFrame의 인덱스를 같게해야 올바른 병합이 일어난다.
org_info_new = pd.concat([org_info, org_agg], axis=1)
org_info_new
```

merge
- 공통된 컬럼을 기준으로 양쪽의 데이터를 병합할 때 사용
- 데이터베이스의 조인 개념과 매우 유사(inner, outer, left, right)

```python
# 급여 지급 내역 데이터 로딩
hr_pay_info = pd.read_csv('hr-pay-info.csv', encoding='euc-kr')
hr_pay_info
```

```python
print(hr_info.shape)
print(hr_pay_info.shape)
```

```python
set(hr_info.columns) - (set(hr_info.columns) - set(hr_pay_info.columns))
```

```python
# 둘 다 가지고 있는 EID컬럼을 중심으로 merge

# 사원 정보 데이터에서 사용할 EID, 이름, 성별
hr_info_df = hr_info[['EID','ENAME','GENDER']]
hr_info_df
```

```python
# 사원 정보 데이터 기준으로 merge
total_salary_df1 = pd.merge(hr_info_df, hr_pay_info,how='left', on='EID')
total_salary_df1
# 병합시, 'EID' 열을 기준으로 병합, 왼쪽 데이터 기준
```

```python
# 급여지급 내역을 중심으로 merge
total_salary_df2 = pd.merge(hr_info_df, hr_pay_info,how='right', on='EID')
total_salary_df2
```

```python
# 급여지급 내역을 중심으로 merge 했을 때 더 적음
# -> 급여지급이 안된 인원이 있다.
```

```python
# 지급 내역이 없는 인원들 확인
total_salary_df1['PAYMENT_DY'].isnull() # null인지 확인
```

```python
# 불리언 인덱싱
no_money = total_salary_df1[total_salary_df1['PAYMENT_DY'].isnull()]
no_money
```

```python
# 지급 내역이 없는 사람들이 언제 퇴사를 했는지 알아보자
pd.merge(no_money,hr_info)[['ENAME','RETIRE_DATE']]

# 퇴사자여서 급여지급이 없었다.
```

### 2. 정렬


sort
- 인덱스를 기준으로 정렬하기 : df.sort_index()
- 값을 기준으로 정렬하기 : df.sort_values()

```python
hr_info.set_index('EID').sort_index()
```

```python
# 컬럼 정렬
hr_info.sort_index(axis=1)
# 알파벳 순
```

```python
# 값을 기준으로 정렬
hr_info.sort_values('SALARY', ascending=False) # 내림차순 정렬
```

```python
# 두가지 기준으로 정렬 -> 나이, 연봉

hr_info.sort_values(['AGE','SALARY'])
# 1. 나이 순으로 오름차순
# 2. 나이가 같은 경우, 그 안에서 연봉으로 오름차순
```

```python
hr_info.sort_values('AGE',inplace=True, ascending=False)
```

### rank

```python
# 랭킹을 매겨주는 함수
# 낮은 값이 rank 1
org_info_new['SALARY'].rank()
```

```python
# 높은 값이 rank 1 : ascending
# 랭킹 값이 들어갈 새로운 열을 추가
org_info_new['SALARY_RANK'] = org_info_new['SALARY'].rank(ascending=False)
org_info_new.head()
```
