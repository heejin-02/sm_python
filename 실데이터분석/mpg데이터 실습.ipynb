```python
# 카테고리별 평균
# groupby
# groupby는 집계함수와 같이 사용됨

# Q1. category별 cty 평균
cty_mean = mpg[['category', 'cty']].groupby('category').mean()
                        # 카테고리별         평균
cty_mean
```

```python
# Q2. 카테고리별 평균을, cty 평균이 높은 순으로 (내림차순)
cty_mean.sort_values('cty',ascending=False)
```

```python
# Q3. hwy 평균이 가장 높은 회사 세 곳 출력
# 제조사, 고속도로연비 데이터만 / .groupby(제조사별) 평균 / 내림차순 / 상위 3개
# 회사별 hwy 평균 구하기 (mean)
hwy_mean = mpg[['manufacturer','hwy']].groupby('manufacturer').mean()
# 평균에서 가장 높은 순으로 정렬
hwy_mean = hwy_mean.sort_values('hwy', ascending=False)
# 높은 회사 세곳 출력 (head)
hwy_mean.head(3)
```

```python
# Q4. 가장 많이 생산한 회사별 'compact' 차종 수 내림차순 정렬

# groupby('제조사별').mean()
# groupby('제조사별').count()

'''
1. category == "compact"인 데이터
2. 에서 manufacturer, category 데이터
3. 에서 groupby('제조사별').count()
4. 에서 category(개수) 별로 내림차순
'''

# 회사별 compact 차종 생산 여부
manu = mpg[['manufacturer', 'category']]
manu[mpg['category']=='compact'].groupby('manufacturer').count().sort_values('category',ascending=False)
```

#### 특정 제조사들의 데이터 살펴보기

```python
# 제조사
mpg['manufacturer'].value_counts()
```

```python
# 방법 1
# 제조사가 toyota인 데이터
toyota = mpg[mpg['manufacturer'] == 'toyota']

toyota
```

```python
ford = mpg[mpg['manufacturer'] == 'ford']
audi = mpg[mpg['manufacturer'] == 'audi']
hyundai = mpg[mpg['manufacturer'] == 'hyundai']
honda = mpg[mpg['manufacturer'] == 'honda']
```

```python
# 제조사가 toyota, ford, audi, hyundai, honda
total = pd.concat([toyota, ford, audi, hyundai, honda], axis = 0)
total
```

```python
# 방법2
# 제조사가 toyota. ford, audi, hyundai, honda
# 조건식
# or -> |
mpg[(mpg['manufacturer'] == 'toyota') | (mpg['manufacturer'] == 'ford') | (mpg['manufacturer'] == 'audi') | (mpg['manufacturer'] == 'hyundai') | (mpg['manufacturer'] == 'honda')]
```

```python
# 방법 3
# 제조사가 toyota, ford, audi, hyundai, honda
# 조건식
# isin : 포함 여부를 판단하는 함수
mpg['manufacturer'].isin(['toyota', 'ford', 'audi', 'hyundai', 'honda'])
```

```python
# 방법4
# manufacturer 열을 인덱스
mpg_manu = mpg.set_index('manufacturer')
mpg_manu
```

```python
# 제조사가 toyota, ford, audi, hyundai, honda
total4 = mpg_manu.loc[['toyota', 'ford', 'audi', 'hyundai', 'honda'],:]
total4
```

```python
# 인덱스 다시 번호로
# reset_index() : 기본 숫자형 인덱스로 변환
total4.reset_index(inplace= True)
total4
```

### 5개 제조사별로 생산하는 자동차 종류를 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns # 시각화 도구
```

```python
total['category'].value_counts()
```

```python
# sns.countplot(data = , x = , hue = ) 
# y값은 알아서 개수가 구해져서 들어간다
# x : x값
# hue : 분할 기준(나타내고자 하는 것)

sns.countplot(data = total, x= 'manufacturer', hue = 'category')
# x축 : 제조사
# 분할기준 : 자동차 종류 (종류별로 분할해서 개수를 나타내라)
```

```python
# 만약 데이터가 결측치가 있을 경우 참고하기 좋다.
# 결측치는 일반적으로 평균, 중앙값, 최빈값을 넣을 수 있지만,
# 시각화 데이터 분석을 참고하여 채울 수 있다.

# 자동차 종류에 결측치가 있다고 가정,
# 1. 제조사가 audi -> compact일 확률이 높다.
# 2. 제조사가 honda일 경우, subcompact일 확률이 높다.
```

### 여러 기준으로 groupby

```python
# 제조사와 자동차 종류별로 고속도로 연비의 평균을 구해보자
# manufacturer, category, hwy
# groupby, mean()

total[['manufacturer', 'category', 'hwy']].groupby(['manufacturer', 'category']).mean()
```

```python
# 구동방식과 연료 종류 별로 고속도로 평균연비와 평균 도시연비
# 필요한 데이터 : drv, fl, hwy, cty
total[['drv', 'fl','hwy','cty']].groupby(['drv','fl']).mean()

# drv
# 4 : 사륜구동
# f : 전륜구동
# r : 후륜구동

# p : 휘발유
# r : 경유
# c : 가스
```

####  제조사별, 자동차 종류별, 고속도로 연비 시각화
- violinplot으로 확인해보자
- violinplot : 수치형 데이터의 '분포'를 확인할 때 사용하는 그래프

```python
# 바이올린 형태로 데이터의 분포와 밀도를 곡선으로 표현
# 데이터가 많이 존재 -> 두껍다
# 데이터가 적게 분포 -> 얇다

# 제조사와 자동차 종류에 따른 고속도로 연비

plt.figure(figsize =(15,6))
sns.violinplot(
    data = total,
    x = 'manufacturer',
    y = 'drv',
    hue = 'year', # 분류 기준(나타내고자 하는 것)
    inner = 'stick' # 바이올린 내부에 개별 데이터선
)
#plt.ylim(0,50)
plt.show()
```
