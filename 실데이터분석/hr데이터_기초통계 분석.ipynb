### 교육 수준에 따른 평균 연봉을 확인

```python
hr_info['EDU_LEVEL'].unique()
```

```python
# 박사 학위 취득자의 평균 연봉을 알아보자
hr_info[hr_info['EDU_LEVEL'] == '박사 학위']['SALARY']. mean()
```

```python
# 교육 수준별 평균 연봉 출력 (반복문)
for edu_level in hr_info['EDU_LEVEL'].unique():
    print(edu_level, ':', hr_info[hr_info['EDU_LEVEL'] == edu_level]['SALARY'].mean())
```

### 3. 데이터 그룹화를 이용한 통계 분석
- groupby

```python
# 교육수준별 연봉
hr_info[['SALARY','EDU_LEVEL']].groupby(by='EDU_LEVEL').mean()
```

그룹화 후 여러 통계값 확인

```python
# 교육 수준과 성별에 따른 //// 연봉의 평균, 최대, 최소, 중앙값 확인
# agg : 여러 집계함수를 한번에 사용할 때
hr_info[['SALARY', 'EDU_LEVEL', 'GENDER']].groupby(['EDU_LEVEL', 'GENDER']).agg(['mean','max','min','median'])
```

#### 그룹화 후 컬럼별로 다른 연산

```python
# 국적에 따른 평균 성과평가점수와 최대 연봉
# 국적에 따른 / '평균' 성과평가점수 / '최대' 연봉

hr_info.groupby('NATION').agg({'PERFORMANCE_SCORE': 'mean',
                               'SALARY' : 'max'})
```

#### 4.  상관계수 분석
- 두 변수 사이의 관계를 수치화하여 보는 방법

#### 분산
- 확률변수가 기댓값으로부터 얼마나 떨어진 곳에 분포하는지 가늠하는 숫자
- 편차 제곱의 평균
#### 표준편차
- 분산의 제곱근

공분산
- 하나의 변수가 아닌, 두개의 변수로 구한 분산값
- 두 변수 사이에 상관관계를 알고싶을 때 사용

```python
# 공분산 값의 문제
# 공분산은 데이터의 스케일에 영향을 받음

# ex) 키(cm), 몸무게(kg)의 공분산 200
# ex) 시력, 눈의 굴절률의 공분산 2

# 공분산을 정규화 -> 상관계수
```

```python
# 상관계수가 1에 가깝다 : 강한 선형관계, A변수 증가하면 B변수도 증가 (양의 선형 상관관계)
# 상관계수가 -1에 가깝다 : 강한 선형관계, A변수가 증가하면 B변수 감소 (음의 선형 상관관계)
```

```python
# 직원 성과 평가 점수와 연봉의 상관계수 확인
hr_info[['PERFORMANCE_SCORE', 'SALARY']].corr()
```

```python
# 수치 데이터들의 상관계수 확인
hr_info[['PERFORMANCE_SCORE','SALARY','AGE','SERVICE_YEAR', 'TRAINING_HOURS']].corr()
```

```python
# 연봉과 나이 / 연봉과 근속년수는 강한 양의 선형관계가 있음
# 나이와 근속연수가 클수록 연봉이 크다
```
