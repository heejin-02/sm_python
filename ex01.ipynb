### 문자열 포매팅
- 문자열 안의 특정한 값을 바꿔야하는 경우

```python
s1 = '오늘은 3월 25일 입니다.'
s1 = '오늘은 3월 26일 입니다.'
print(s1)
```

```python
# % 기호를 사용하는 방법
day = 22
# 정수형 변수를 넣을 자리에 %d, 마지막에 들어갈 %들어갈 변수명
s1 = '오늘은 3월 %d일 입니다.'%day

# %d : 정수형
# %f : 실수형
# %s : 문자형

print(s1)
```

```python
day += 1
# day = day + 1
print('오늘은 3월 %d일 입니다.'%day)
```

```python
# format 함수를 사용한 포매팅
month = 3
day = '이십오'

# .format(첫번째 변수, 두번째 변수,...)
# 자료형이 달라져도 상관없다
s = '오늘은 {}월 {}일 입니다.'.format(month,day)

print(s)
```

```python
# f문자열을 사용한 포매팅
# 현재 가장 많이 쓰이는 방법. python3 부터 제공하는 기능

month = 4
day = 5

s = f'오늘은 {month}월 {day}일 입니다.'
print(s)
```
