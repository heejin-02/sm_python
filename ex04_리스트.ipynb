```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

#2 출력
print("첫 번째 요소의 두 번째 요소:",matrix[0][1])
#[4,5,6] 출력
print("두번째 요소 전체:",matrix[1])
#[[4,5,6],[7,8,9]] 를 출력
print("두 번째 요소부터 끝까지:",matrix[1:])
# [4,5]를 출력해라
print("두 번째 요소에서, 첫 번째부터 두 번째 요소까지 :",matrix[1][:2])
```

```python
# list의 덧셈 연산
list4 = [1,2,3]
list4_2 = [3,4,5]

list4+list4_2
```

```python
# list의 곱셈 연산
list4*4
```

```python
# list의 요소 변경 및 추가

list5 = [0,1,2]
print('변경 전', list5)

#요소 변경
list5[2] = 5
print('변경 후', list5)

#리스트의 요소 추가
# append
list5.append(10)
print('추가 후', list5)
```
