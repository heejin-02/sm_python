```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

#2 출력
print("첫 번째 요소의 두 번째 요소:",matrix[0][1])
#[4,5,6] 출력
print("두번째 요소 전체:",matrix[1])
#[[4,5,6],[7,8,9]] 를 출력
print("두 번째 요소부터 끝까지:",matrix[1:])
# [4,5]를 출력해라
print("두 번째 요소에서, 첫 번째부터 두 번째 요소까지 :",matrix[1][:2])
```

```python
# list의 요소 변경 및 추가

list5 = [0,1,2]
print('변경 전', list5)

#요소 변경
list5[2] = 5
print('변경 후', list5)

#리스트의 요소 추가
# append
list5.append(10)
print('추가 후', list5)
```

```python
#슬라이싱으로 요소 변경
list5 = [0,1,2,3,4,5,6]
print('변경 전 : ',list5)

list5[2:5] = ['a','b','c',11,22] # 리스트의 형태로 넣어주어야한다.
print('변경 후 : ',list5) 
```

```python
array = [1,2,3,4,5]
#array 5 -> 7 변경
array[4] = 7
print(array)

#array 2 -> 리스트['a','b','c']로 변경
array[1] = ['a','b','c']
print(array)
# 인덱싱으로 값을 변경하는 경우
# 해당 요소에 요소 자체를 그대로 변경

#array 4,7을 'd','e','f','g'로 변경
array[3:] = 'd','e','f','g'
print(array)
# 슬라이싱으로 값을 변경하는 경우
# 변경할 값(리스트)에 담긴 값들의 각각의 요소로 변경
```

```python
fruits = ['사과','포도',['수박','멜론'],'오렌지']
apple = fruits[0]
melon = fruits[2][1]
orange = fruits[3]

choice_lst =[]
choice_lst.append(apple)
choice_lst.append(melon)
choice_lst.append(orange)
choice_lst
```

```python
# 한번에 값을 넣어주는 방법
choice_lst2 = []
# extend : 리스트 안에 여러개 값을 한번에 추가
choice_lst2.extend([apple,melon,orange]) # list의 형태로 넣어줌
choice_lst2
```

### 리스트에 사용되는 다양한 함수들

```python
# 길이(요소 개수)를 나타내는 함수
#len

lst = [2,3,4,5,6,6,2,23,4,25,2,6,443,53,32,4,4,5,2,1]

len(lst)
```

```python
len('안녕하세요') #문자열의 길이도 알 수 있다.
```

```python
# 특정 값의 인덱스 반환
# index()
# list.index(찾고자 하는 값)
lst.index(2) # 중복되면 맨 처음에 있는 값 인덱스가 나온다.
```

```python
# 값을 추가하는데, 끝이 아닌 원하는 위치에 추가
# insert
# list.insert(인덱스 위치, 값 )
list5 = [0,1,2,3,4]
print('insert 전 : ', list5)

list5.insert(2, '추가된 값')
print('insert 후 : ',list5)
```

```python
# list 값 삭제

list7 = [33,44,55,66,77]
print('삭제 전 :', list7)

#인덱스 번호로 삭제
# del
del list7[3]
print('삭제 후 :', list7)

# 값으로 삭제
# list.remove(삭제할 값)
list7.remove(44)
print('44 삭제 후 :',list7)
```

```python
# 값이 여러개일 경우 remove 함수
list8 = [1,5,1,4,3,5,9,9,9,9,9,2,1,5]
print('삭제 전 : ',list8)

# 9 삭제
list8.remove(9) #가장 처음 있는 9만 삭제
print('삭제 후 : ',list8)
```

```python
# 9 삭제
list8.remove(9) #선언과 따로 실행하면 실행할때마다 9가 삭제된다.
print('삭제 후 : ',list8)
```

```python
# 슬라이싱을 활용해서 삭제
list7 = [33,44,55,66,77]
print('삭제 전 : ', list7)

# del
del list7[1:3]
print('삭제 후 : ', list7)
```

```python
# 오름차순 정렬
# sort()
# list.sort()

list8 = [3,9,7,3,4,5]
list8.sort()
list8
```

```python
# 역순
# reverse()
# list.reverse()

list8 = [3,9,7,3,1,5,6]

list8.reverse()
list8
```

```python
# 내림차순
list8 = [3,9,7,3,1,5,6]
list8.sort() # 정렬
list8.reverse() # 후 역순으로 바꾸기

list8
```
