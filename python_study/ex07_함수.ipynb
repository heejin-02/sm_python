- 호출될 때, 특정 기능을 실행하는 재사용 가능한 코드 블록

### 함수의 목적
- 반복적인 프로그래밍을 피할 수 있다.
- 모듈화로 인해 코드의 가독성이 좋아진다.
- 프로그램에 문제가 발생하거나, 기능의 변경이 필요할 때, 손쉽게 유지보수가 가능하다.

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

# 빼기 기능을 하는 함수 호출
result = number_sub(num1,num2)
print(result)
```

```python
# number_sub 라는 함수를 정의
def number_sub(num1,num2): # 함수를 정의, 매개변수로는 n1, n2 사용
    result= num1-num2       # 실행문장, result라는 변수에, 매개변수끼리 뺀 값을 저장
    return result # result 변수의 값을 반환
```

```python
def s_replace(str1):
    result = ''  # 새로운 문자열을 저장할 변수
    for i in str1:  # 원래 문자열을 순회
        if i != 'ㅋ':  # 'ㅋ'이 아닐 때만 추가
            result += i
    return result
```

```python
# replace()
# str.replace(바뀔 문자, 바꿀 문자)
str1 = '콩국수에는 설탕을 넣어야 제맛이다.'
print('바꾸기 전 :',str1)
print('바꾼 후: ' ,str1.replace('설탕','소금'))

                            
```

```python
def divisor(num):
    '''
    약수를 구해서 출력해주는 함수
    ''' 
    for i in range(1,num+1):
        if num % i == 0:
            print(i, end=' ')
```

```python
divisor(10)
```
