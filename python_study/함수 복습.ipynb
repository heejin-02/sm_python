```python
s = input('문자열 입력 >>')
result = s_replace(s)
result
```

```python
def s_replace(s):
    result = '' # result라는 문자형 변수 선언
    for i in s:
         # s을 정의하지 않았지만, 인수를 받아오면서 정의
            # result라는 빈 문자열이 담겨있는 변수에 문자를 하나씩 추가
            # 각 문자가 'ㅋ'이 아닌 경우에만
        if i != 'ㅋ':# 문자가 'ㅋ'이 아니라면
            result += i # result라는 변수에 해당 문자를 추가
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")
result = cal(num1, num2, op)
print(f"결과 : {result}")
```

```python
def cal(num1, num2, op):
    if op == '+':
        return num1+num2
    elif op == '-':
       return num1 - num2
    else:
       return "잘못된 입력입니다"
```

### aply 함수
- 각 요소에 특정 함수를 적용시켜주는 함수

```python
import pandas as pd
```

```python
pop = pd.Series([9600000,3340000, 1480000, 2410000],
               index = ['서울','부산','광주','대구']
               )
pop
```

```python
pop + 100
```

```python
# 값을 100씩 더해주는 함수 정의
def plus_100(x):
    return x+100
```

```python
# apply 함수 사용
# Series의 개별적 요소에 적용
pop.apply(plus_100)
```

```python
# 함수에 인수로 pop를 넣어보자
plus_100(pop)

# pop가 함수에 들어가서
# return 값: pop + 100
```

```python
# 차이점
# 1. plus_100(pop) : pop 자체를 함수의 인수로서(입력값) 전달 -> 반환값 : pop + 100
# 2. pop.apply(plus_100) : 함수가 각각 pop 요소에 개별적으로 접근
# -> 9600000 + 100, 3340000 + 100 ... 
```

```python
# 300만 이상인 값만 표시해주는 함수(그 외에는 0 반환)
def con_return(x):
    if x >= 3000000:
        return x
    else:
        return 0
```

```python
# apply 함수 사용
pop.apply(con_return)
```

```python
# apply 미사용
# 함수 안에 pop 인수로 대입
con_return(pop)
```

```python
# lambda x : 익명함수 정의(잠깐 사용할 함수)
# apply 함수와 자주 사용
pop.apply(lambda x : x+200)
```

```python
# 숫자에 천 단위 구분 기호를 추가
pop.apply(lambda x : f'{x:,}')
# 문자열 포매팅 f'{x: }' x라는 변수에 특정 포매팅 적용 (,는 천단위 구분하라는 기호)
```
