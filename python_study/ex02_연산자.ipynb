```python
py_score = int(input('python 점수 입력 >>'))
ml_score = int(input('머신러닝 점수 입력 >>'))
dl_score = int(input('딥러닝 점수 입력 >>'))
score_sum = py_score + ml_score + dl_score
score_avg = (py_score + ml_score + dl_score) / 3
print('합계 : ',score_sum)
print('평균 : ',score_avg)
```

```python
time = int(input("시간(초) 입력 >> "))
hour = (time // 3600) # 3600초 = 1시간, 3600으로 나눈 몫
minute = (time % 3600) // 60 # (시간)으로 빠진 부분은 제외, 남은 초를 계산
second = (time % 3600) % 60 # (분)으로 빠진 부분까지 제외, 남은 초를 계산
print(f"{hour}시간 {minute}분 {second}초")
```

### 비교 연산자
- boolean 값 반환(True, False)
  

```python
a = 3
b = 7
print(a > b)
print(a >= b)
print(a == b)
print(a != b)

print()

# 논리 연산자 not : 논리값을 뒤집음
print(not a > b)
print(not a >= b)
print(not a == b)
print(not a != b)
```

### 멤버 연산자 in

```python
s = '콩국수에 들어갈건 소금뿐'

# 조건식 입력
'소금' in s
# '소금'이라는 문자가 s 안에 있니?
```

```python
s = '콩국수에 들어갈건 소금뿐'

# 조건식 입력
'소금' not in s
# '소금'이라는 문자가 s 안에 없니?
```

### 삼항 연산자
- 조건문의 단순화
- a if 조건식 else b


```python
# 삼항연산자
java_score = int(input('자바 점수 입력 : '))
'합격' if java_score >=60 else "불합격"
# 합격이다. 만약에, java 점수가 60점 이상이라면 합격, 그렇지않으면 불합격
```

```python
num = int(input('정수 입력 >> '))
f'{num}는(은) 짝수입니다.' if num % 2 ==0 else f'{num}는(은) 홀수입니다.'
```
