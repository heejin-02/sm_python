### 조건문
- 상황에 따라 실행 흐름을 다르게 하는 문법
- 키워드 : if, elif, else
- 콜론(:)과 들여쓰기(tap) 문법이 중요

```python
# 들여쓰기로 조건문의 실행문장을 구분한다.
if True:
    print('희진씨는 팀장이다')
print('이 문장은 조건문 밖에 있는 문장이라 항상 실행된다')
```

```python
if False:
    print('설탕국수는 존맛탱이다.')
print('이 문장은 조건문 밖에 있는 문장이라 항상 실행된다') 
```

```python
# 간단한 if 조건문
money = 9000
if money >= 10000:
    print('택시를 탄다.')
else:
    print('버스를 탄다.')
```

```python
num = int(input('정수 입력 >> '))
if num % 3 ==0 and num % 5 == 0:
    print('3과 5의 배수입니다.')
else:
    print('3과 5의 배수가 아닙니다.')
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
if num1 > num2:
    print('첫번째 정수가 더 큽니다.')
elif num1 < num2:
    print('두번째 정수가 더 큽니다.')
else:
    print('두 수가 똑같습니다.')
```

```python
score = int(input('점수 입력 >> '))
if score <= 100 and score >= 0:
    if score >= 90:
        print(f'{score}점은 A학점 입니다.')
    elif score >= 80:
        print(f'{score}점은 B학점 입니다.')
    elif score >= 70:
        print(f'{score}점은 C학점 입니다.')
    elif score >= 60:
        print(f'{score}점은 D학점 입니다.')
    else:
        print(f'{score}점은 F학점 입니다.')
else:
    print('잘못된 입력입니다.')
```

```python
score = int(input('점수 입력 >> '))
if score <= 100 and score >= 0:
    if score >= 90:
        grade ='A'
    elif score >= 80:
        grade ='B'
    elif score >= 70:
       grade ='C'
    elif score >= 60:
       grade='D'
    else:
       grade='F'
    print(f'{score}점은 {grade}학점 입니다.')
else:
    print('잘못된 입력입니다.')
```

```python
x = int(input('x 입력 >> '))
y = int(input('y 입력 >> '))
if x != 0 and y !=0:
    if x >0 and y>0:
        qua = 1;
    elif x<0 and y>0:
        qua = 2;
    elif x>0 and y<0:
        qua = 4;
    else:
        qua = 3;
    print(f'({x},{y})는 {qua}사분면 입니다.')
elif x== 0 and y==0:
    print('원점입니다')
else:
    print('잘못된 입력입니다.')
```
