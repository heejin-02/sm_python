```python
all_stock = req.get("https://finance.naver.com/sise/sise_market_sum.naver")
```

```python
# 데이터를 특정 규칙에 맞게 구조화된 형태로 바꿈
soup = bs(all_stock.text,"lxml")
```

```python
# HTML 데이터에서 특정 요소(태그, 콘텐츠) 수집
stock_title = soup.select("a.tltle")
price = soup.select("table.type_2>tbody>tr>td:nth-child(3)") 
```

```python
# 순수 텍스트만 남게 변형
price_list = []
for i in range(len(price)):
    price_list.append(price[i].text)
print(price_list)
```

```python
title_list = []
for i in stock_title:
    print(i.text)
    title_list.append(i.text)
    
```

```python
title_list
```

```python
# csv 형태로 저장
import pandas as pd
```

```python
dic = {"종목명" : title_list, "가격" : price_list}
stock = pd.DataFrame(dic)
stock.to_csv("시가총액.csv", encoding = "euc-kr")
```
