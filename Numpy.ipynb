#### 배열끼리의 사칙 연산

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr3 = np.array([7,8,9,10])
```

```python
arr1 + arr2
```

```python
arr1+ arr3 # 원소 구조가 동일하지 않아 오류
```

```python
print(arr2 + arr1)
print(arr2 - arr1)
print(arr2 * arr1)
print(arr2 / arr1)
print(arr2 // arr1)
print(arr2 % arr1)
```

## 편리한 배열 생성 함수

```python
# 0으로 배열 생성
# zeros((형태))

arr_0 = np.zeros((3,2))
arr_0
```

```python
# 1로 배열 생성
# ones
arr_1 = np.ones((3,2), dtype=np.int32)
arr_1
```

```python
# 특정 값으로 배열 생성
# full
# np.full((형태),값)
arr_f = np.full((2,3,2),1.0)
arr_f
```

```python
# np.arange():range() 시퀀스로 배열 생성, 사용법은 range() 함수와 유사
# np.arange(시작할 값, 끝값(포함하지 않음, 증감량))

arr_a = np.arange(1,13,1)
print(arr_a)

arr_a = np.arange(13) # 0부터 12까지
print(arr_a)
```

```python
arr_a = np.arange(12)
print(arr_a)

# reshape : 배열의 형태(shape) 변경
arr_a.reshape(3,4) # 3행 4열의 새로운 배열로 0부터 11까지 생성
```

### Boolean 인덱싱
- 특정 조건에 맞는 값들을 찾아내고 싶을 때
- 사용법 : 인덱스 번호 대신, '조건식' 입력

```python
# 조건식
a = 5

a > 3
```

```python
type(a<3)
```

```python
arr5
```

```python
arr5 >= 70
```

```python
type(arr5 >= 70)
```

```python
# 다중조건
# 65 이상이고, 90이하

# numpy에서는 and or 사용 불가
# numpy에서는 &, |를 사용, numpy 배열의 내부코드가 비트연산을 기반으로 만들어져있음
# & : and | : or 연산자   

# arr5 >= 65 and arr5 <= 90
(arr5 >= 65) & (arr5 <= 90)
```

```python
# 짝수 값 판단 조건식
arr5 % 2 == 0 

# 불리언 인덱싱을 위한 재료
```

```python
# 실제 값을 가져와보자
# True에 해당하는 값들을 가져와보자
# 인덱스 번호대신, '조건식' 입력
```

```python
arr5[arr5 % 2 == 0]
```

```python
arr6 = np.array([[6,9,6,8,7],
                 [2,3,5,4,1]])
arr6
```

```python
# 짝수일 경우, 0으로 변경하는 코드를 작성
# 불리언 인덱싱으로 값에 접근해서, 0을 대입
arr6[arr6 % 2 == 0] = 0
arr6
```

```python
# 조건식
arr6 % 2 ==0
```

### 평균과 합계를 구하는 함수

```python
arr7 = np.array([[1,2,3],[4,5,6]])
arr7
```

```python
# 합계
# arr.sum()
arr7.sum()
# 배열이라는 객체에 속한 메서드
```

```python
sum(arr7) # 내장함수 sum : 같은 배열끼리만 더함
```

```python
sum(sum(arr7)) # 배열 내 모든 수를 또 더함
```

```python
# numpy 라이브러리 속 sum 함수
# np.sum(시퀀스형 자료구조)
print(np.sum(arr7))
print(np.sum([1,2,3,4,5,6]))
print(np.sum(range(1,7)))
```

```python
# 평균
# arr.mean()
arr7.mean()
```

```python
# numpy 라이브러리 속 mean 함수
# np.mean(시퀀스형 자료구조)
print(np.mean(arr7))
print(np.mean([1,2,3,4,5,6]))
print(np.mean(range(1,7)))
```
