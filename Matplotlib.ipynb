## Matplotlib
#### Mathdatas plot library
- 메타데이터를 도표나 그래프(시각화)로 출력시키는 라이브러리
- 메타데이터 : 일정한 규칙에 따라 구조화 표준화된 데이터
- 분석 결과를 수치만으로 설명하는 것보다 더 직관적으로 이해 가능
- 타인에게 분석 결과를 전달하기에 효과적

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

### 플롯 차트(plot chart) : 선 그래프

```python
y = [2,4,5]

plt.plot(y) # 선 그래프
plt.show()  # 콘솔에 차트 출력
```

```python
# x, y 값 지정하여 선 그래프
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls = '-', lw=2, c = 'red', marker = 'o', mfc = 'yellow', mec = 'blue', ms = 5)
# ls : 선 스타일
# lw : 선 굵기
# c : 선 색깔
# marker : 마커의 종류 
# mfc : 마커의 색깔
# mec : 마커 테두리 선 색깔

plt.show()
```

### 실습

```python
x = np.arange(7)
y = np.arange(7)

plt.plot(x,y ,c = 'gold',marker ='*', ms=10, mec ='yellow', mfc = 'yellow')
plt.fill(x, y, 'green', alpha=0.5) 
plt.axis("equal")  # 축 비율을 동일하게 설정
plt.axis("off") 
plt.show()
```

### 그리드
- 차트의 선 위치들을 잘 보여주기 위해 차트 중간에 격자 선을 그어주는 것

```python
x = np.arange(7)
y = [1,5,3,12,5,67,2]
plt.plot(x,y ,c = 'gold',marker ='*', ms=10, mec ='yellow', mfc = 'yellow')
plt.grid
plt.show()
```

#### 틱 설정
- 플룻이나 차트에서 축 상의 위치 표시 지점 : 틱(tick)
- 자동으로 설정, 수동설정 가능

```python
x = np.arange(7)
y = [1,15,5,5,5,15,1]
plt.plot(x,y ,c = 'gold',marker ='s', ms=10, mec ='brown', mfc = 'brown')

plt.xticks([1,5,1])
plt.yticks([0,3,6])

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('CAT')
plt.grid # 차트 내 격자 생성
plt.show()
```

x,y 축 그래프 제목 설정

### 여러개 플롯 한꺼번에 그리기

```python
x = np.arange(4)
y1 = [1,1,5,10]
y2 = [1,4,6,8]
y3 = [1,2,4,5]
plt.plot(x,y1, label = 'y1')
plt.plot(x,y2, label = 'y2')
plt.plot(x,y3, label = 'y3',c= 'green')

#범례 설정
plt.legend(loc = 'upper left',prop = {'size':10})
# loc : 범례의 위치 (upper, center, lower) (left, center, right)
# prop : 범례의 크기


plt.grid # 차트 내 격자 생성
plt.show()
```

### 바 차트(bar chart): 막대 그래프

```python
x = np.arange(6)
y = [70,70,50,50,70,70]

# bar chart 만들어주는 함수
plt.bar(x,y, color = ['pink'], edgecolor ='pink', linewidth = 3,
       alpha = 1, width = 1)

# color : 바 내부 색상
# edgecolor : 테두리 색상
# linewidth : 테두리 너비
# alpha : 투명도(0~1, 기본값 1)

plt.show()
```

### 주간에 일어난 자동차끼리의 사고를 요일별로 시각화
- 사고유형_대분류 : 차대차
- 주야 : 주간
- 요일별로 시각화

```python
data['사고유형_대분류'].value_counts()
```

```python
# 사고유형 대분류가 '차대차'
# 불리언 인덱싱

# 조건식
data['사고유형_대분류'] == '차대차'
```

```python
car_car = data[data['사고유형_대분류'] == '차대차']
car_car
```

```python
# 데이터 프레임 모든 열(컬럼) (일부 열 가능) 출력
# (None : 모든 열 출력, 숫자 기입 시 숫자만큼)
pd.set_option('display.max_columns',None)

car_car
```

```python
# 사고 유형이 차대차 이면서, (car_car) 주야가 '주간'인 데이터
#불리언 인덱싱

#조건식
car_car['주야'] == '주간'
```

```python
noa = car_car[car_car['주야'] == '주간']
noa
```

```python
# 요일별로 사건 발생 횟수
date = noa['요일'].value_counts()
#noa.value_counts('요일')
date
```

```python
# 요일 순서 정렬
date = date[['월','화','수','목','금','토','일']]
```

```python
# 차트 내 폰트 변경을 위한 도구
from matplotlib import rc
rc('font',family = 'Malgun Gothic')
```

```python
date.values
```

```python
y = date.values # y값
x = date.index # x축

plt.bar(x,y)

plt.ylim(0,180)
plt.title('요일별 주간 차대차 교통사고 건수') # 그래프 제목
plt.xlabel('요일') # x축 라벨
plt.ylabel('사고 건수') # y축 라벨

plt.show()
```
